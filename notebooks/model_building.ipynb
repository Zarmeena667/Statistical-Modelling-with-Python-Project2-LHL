Build a regression model.

```python
import numpy as np
import pandas as pd
from sklearn import linear_model, datasets
import statsmodels.api as sm
```

```python
df= pd.read_csv('finalfinal_df.csv')
df
```

```python
X= df['Distance']
Y= df['Rating']
```

```python
X = sm.add_constant(X)
lin_reg = sm.OLS(Y,X)
```

```python
model = lin_reg.fit()
print_model = model.summary()
print(print_model)
```

Provide model output and an interpretation of the results. 

The merged dataframe has a high number of null values and imputed values (Ratings and Distance). It is evident from the supremely low R-squared value of 0.008 that this is not the best fit model. 
A p-value of less than 0.05 is considered to be statistically significant. This regression output shows a p-value of 0, which means that the distance does not impact the rating.
