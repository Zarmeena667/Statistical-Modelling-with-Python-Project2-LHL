# CityBikes

Send a request to CityBikes for the city of your choice. 

```python
import requests
import pandas as pd
import os
import json
```

```python
response = requests.get(url = "http://api.citybik.es/v2/networks/biketown")
response.json()
```

Parse through the response to get the details you want for the bike stations in that city (latitude, longitude, number of bikes). 

```python
response.json().keys()
```

```python
response.json()['network'].keys()
```

```python
response.json()['network']['stations']
```

Put your parsed results into a DataFrame.

```python
stations = response.json()['network']['stations']
df= pd.json_normalize(stations)
df
```

```python
#dropping irrelevant columns
cb= df.drop([ 'timestamp', 'extra.has_ebikes', 'extra.last_updated', 'extra.rental_uris.android', 'extra.rental_uris.ios',	'extra.renting'	,'extra.returning',
             	'extra.slots', 'extra.uid', 'extra.ebikes'], axis=1)
cb
```

```python
#renaming columns
rename_cb = cb.rename(columns= {'id':'ID', 'latitude': 'Latitude', 'longitude':'Longitude', 'name': 'Name', 'extra.address':'Address'})
rename_cb
```

```python
#storing to a CSV file
rename_cb.to_csv('city_bikes_data.csv', index = 'False')
```
